```python
# Lambda function to add two numbers

add = lambda a, b: a + b

print(add(5, 3))
```

```python
# Lambda function to find square of a number

square = lambda x: x * x

print(square(4))
```

```python
# Lambda function to check even or odd

check = lambda x: "Even" if x % 2 == 0 else "Odd"

print(check(7))
```

```python
# Lambda function to find maximum of two numbers

maximum = lambda a, b: a if a > b else b

print(maximum(10, 20))
```

```python
# Lambda function to sort list in ascending order

numbers = [5, 2, 9, 1, 7]

sorted_list = sorted(numbers)

print(sorted_list)
```
